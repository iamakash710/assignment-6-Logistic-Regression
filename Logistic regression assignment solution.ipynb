```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
bank=pd.read_csv('bank-full.csv', sep=';')

bank.head()
```

## EDA

- TO CONVERT THE CHATEGORICAL VALUES TO NUMERIC VALUES USING DUMMY VARIABLES

```python
# to get all the chategorical values converted into the dummy variables
bank_cl1= pd.get_dummies(bank, columns=['job', 'marital', 'education', 'contact','poutcome'])
bank_cl1('max')
```

```python
# to display all the columns
pd.set_option('display.max.columns', None)
bank_cl1
```

```python
bank_cl1.info()
```

- TO CONVERT THE YES/NO VALUES TO NUMMERIC VALUES USING THE BINARY ENCODING METHOD

```python
# to convert all the yes/no type data to numeric values (binary encoding)
bank_cl1['default']=np.where(bank_cl1['default']=='yes',1,0)
bank_cl1['housing']=np.where(bank_cl1['housing']=='yes',1,0)
bank_cl1['loan']=np.where(bank_cl1['loan']=='yes',1,0)
bank_cl1['y']=np.where(bank_cl1['y']=='yes',1,0)
bank_cl1
```

```python
bank_cl1.info()
```

```python
#to check the values in month column
bank_cl1['month'].value_counts()
```

- TO REPLACE THE MONTH VALUES WITH NUMBER LABELS USING REPLACE FUNCTION

```python
seq={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':8,'nov':9,'dec':10}}
bank_cl2=bank_cl1.replace(seq)
bank_cl2
```

- NOW WE HAVE CONVERTED ALL THE CHATEORICAL VALUES TO THE NUMERIC VALUES AND NOW WE ARE READY TO BUILD THE MODEL 

## MODEL BUILDING

```python
x=bank_cl2.drop(['y'], axis=1)
y=bank_cl2['y']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test=train_test_split(x,y,test_size=0.3, random_state=10)
```

## MODEL TRAIN

```python
import warnings
warnings.filterwarnings('ignore')


model=LogisticRegression()
model.fit(x_train,y_train)
```

## MODEL PREDICTIONS FOR THE TRAIN DATA

```python
y_train_pred=model.predict(x_train)
y_train_pred
```

```python
train_df=pd.DataFrame({'y_train':y_train, 'y_train_pred':y_train_pred})
train_df
```

## MODEL PREDICTIONS FOR THE TEST DATA

```python
y_test_pred=model.predict(x_test)
y_test_pred
```

```python
test_df=pd.DataFrame({'y_test':y_test,'y_test_pred':y_test_pred})
test_df
```

## MODEL TESTING


```python
from sklearn.metrics import accuracy_score
```

- ACCURACY FOR THE TRAIN DATA

```python
acc=accuracy_score(y_train,y_train_pred)
print(f'accuracy for the train data is {acc}')
```

- ACCURACY FOR THE TEST DATA

```python
acc_test=accuracy_score(y_test,y_test_pred)
print(f'accuracy for the test data is {acc_test}')
```

#### HERE WE CAN CLEARLY SEE THAT THE TEST ACCURACY AND THE TRAIN ACCURACY ARE PRETTY MUCH THE SAME SO WE CAN USE THIS MODEL FOR NEW PREDICTIONS
